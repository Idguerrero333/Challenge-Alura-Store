### Importación de datos



```python
import pandas as pd # Importa la librería pandas y le asigna el alias 'pd' para facilitar su uso

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv" # Define la URL del archivo CSV correspondiente a la tienda 1
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url) # Lee el archivo CSV desde la URL de tienda 1 y lo carga en un DataFrame llamado 'tienda'
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head() # Muestra las primeras 5 filas del DataFrame 'tienda' para visualizar su contenido
```

#1. Análisis de facturación



##**1.1. Tienda 1**

```python
tienda.groupby('Producto')[['Precio', 'Costo de envío']].sum().sort_values('Precio',ascending=False) # Agrupa el DataFrame 'tienda' por producto y calcula la suma de las columnas numéricas en orden descendente
```

```python
tienda.groupby('Producto')[['Precio', 'Costo de envío', 'Calificación', 'Cantidad de cuotas']].mean(numeric_only=True) # Agrupa el DataFrame 'tienda' por producto y calcula el promedio de las columnas numéricas
```

```python
porcentaje_producto = tienda.Producto.value_counts(normalize=True).to_frame()*100 # Calcula el porcentaje de aparición de cada producto y lo convierte en un DataFrame
porcentaje_producto.plot(kind='bar', figsize=(12,8), color='green', xlabel='Tipos', ylabel='Porcentaje') # Genera un gráfico de barras a partir del DataFrame de porcentajes
```

```python
tienda[['Precio', 'Costo de envío']].sum() # Calcula en una Tupla la suma de la columna 'Precio' y 'Costo de envío' del DataFrame 'tienda'
```

# 4. Productos más y menos vendidos

##**4.1. Tienda 1**
*Productos más y menos vendidos clasificados por top(5) head y tail*

```python
tienda.Producto.value_counts(normalize=True, ascending=False).head()*100 # Calcula el porcentaje de aparición de cada producto en la columna 'Producto' y muestra los primeros 5
```

```python
top_head_t1 = tienda.Producto.value_counts(normalize=True, ascending=False).head()*100 #
top_head_t1.plot(kind='bar', figsize=(12,8), color='green', xlabel='Productos', ylabel='Porcentaje cantidad ventas')
```

```python
top_tail_t1 = tienda.Producto.value_counts(normalize=True, ascending=False).tail()*100
top_tail_t1.plot(kind='bar', figsize=(12,8), color='green', xlabel='Productos', ylabel='Porcentaje cantidad ventas')
```

```python
porcentaje_producto = tienda.Producto.value_counts(normalize=True).to_frame()*100
porcentaje_producto.plot(kind='bar', figsize=(12,8), color='green', xlabel='Productos', ylabel='Porcentaje cantidad ventas')
```
