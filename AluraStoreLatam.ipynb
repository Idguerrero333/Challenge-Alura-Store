Analizando la distribucion geografica de los productos hayamos que las categorias relacionadas se concentran en la latidud 5 y longitud -75

*   Instrumentos musicales
*   Libros
*   Deportes y diversión

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Graficar cada categoría por separado
for categoria in tienda['Categoría del Producto'].unique():
    subset = tienda[tienda['Categoría del Producto'] == categoria]
    plt.scatter(subset['lat'], subset['lon'], label=categoria)

plt.title('Distribución geográfica por Categoría del Producto')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.legend(title='Categoría')
plt.grid(True)
plt.show()
```

Con el siguiente grafico podemos apreciar la densidad de productos vendidos sobre las cordenadas, entre latitud 5 y longitud -74

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.kdeplot(
    data=tienda,
    x='lat',
    y='lon',
    fill=True,        # Rellena el área de densidad
    cmap='Reds',      # Paleta de color para la concentración
    thresh=0.05,      # Umbral mínimo de densidad para mostrar
    levels=100        # Suavidad del gradiente
)

plt.title('Mapa de calor de concentración geográfica')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.grid(True)
plt.show()
```

El siguiente grafico es exportable y muestra en mapa geografico, la concentracion de las ventas, donde se observa mayor concentracion en los envios hacia Bogotá D.C., el eje cafetero y la Costa.

```python
import folium
from folium.plugins import HeatMap

# Crear mapa centrado en el promedio de coordenadas
mapa = folium.Map(location=[tienda['lat'].mean(), tienda['lon'].mean()], zoom_start=12)

# Agregar capa de calor
heat_data = tienda[['lat', 'lon']].dropna().values.tolist()
HeatMap(heat_data).add_to(mapa)

# Mostrar mapa (en Jupyter o exportar como HTML)
mapa.save('mapa_calor_t1.html')
```

##**Tienda 2**

Analizando la distribucion geografica de los productos hayamos que las categorias relacionadas se concentran en la latidud 5 y longitud -75

*   Instrumentos musicales
*   Libros
*   juguetes

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Graficar cada categoría por separado
for categoria in tienda2['Categoría del Producto'].unique():
    subset = tienda2[tienda2['Categoría del Producto'] == categoria]
    plt.scatter(subset['lat'], subset['lon'], label=categoria)

plt.title('Distribución geográfica por Categoría del Producto')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.legend(title='Categoría')
plt.grid(True)
plt.show()
```

Con el siguiente grafico podemos apreciar la densidad de productos vendidos sobre las cordenadas, entre latitud 5 y longitud -74

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.kdeplot(
    data=tienda2,
    x='lat',
    y='lon',
    fill=True,        # Rellena el área de densidad
    cmap='Reds',      # Paleta de color para la concentración
    thresh=0.05,      # Umbral mínimo de densidad para mostrar
    levels=100        # Suavidad del gradiente
)

plt.title('Mapa de calor de concentración geográfica')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.grid(True)
plt.show()
```
