# 4. Productos más y menos vendidos

##**4.1. Tienda 1**
*Productos más y menos vendidos clasificados por top(5) head y tail*

```python
top_head_t1=tienda['Producto'].value_counts(normalize=True).to_frame().sort_values('proportion', ascending=False).head(5)*100 # Calcula el porcentaje de aparición de cada producto en la columna 'Producto' y muestra los primeros 5
top_tail_t1=tienda['Producto'].value_counts(normalize=True).to_frame().sort_values('proportion', ascending=False).tail(5)*100 # Calcula el porcentaje de aparición de cada producto en la columna 'Producto' y muestra los 5 utilmos
top_head_t1=round(top_head_t1,2)
top_tail_t1=round(top_tail_t1,2)
```

```python
top_head_t1.rename(columns={'proportion': 'Porcentaje %'}, inplace=True)
top_tail_t1.rename(columns={'proportion': 'Porcentaje %'}, inplace=True)
top_head_t1['Producto'] = ['Microondas', 'TV LED UHD 4K', 'Armario', 'Secadora de ropa', 'Mesa de noche']
top_head_t1
```

```python
top_head_t1.plot(kind='bar', figsize=(12,8), color='green', xlabel='Productos', ylabel='Porcentaje cantidad ventas')
```

```python
top_tail_t1.plot(kind='bar', figsize=(12,8), color='green', xlabel='Productos', ylabel='Porcentaje cantidad ventas')
```
